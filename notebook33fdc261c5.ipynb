```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
data=pd.read_csv('/kaggle/input/loan-approval-classification-data/loan_data.csv')
```

```python
data
```

```python
mint=data['credit_score'].min()
print(mint)
```

```python
le = LabelEncoder()
label = le.fit_transform(data['person_gender'])
data.drop("person_gender", axis=1, inplace=True)
 
# Appending the array to our dataFrame 
# with column name 'person_gender'
data["person_gender"] = label
```

```python
le1 = LabelEncoder()
label1 = le1.fit_transform(data['previous_loan_defaults_on_file'])
data.drop("previous_loan_defaults_on_file", axis=1, inplace=True)
 
# Appending the array to our dataFrame 
# with column name 'person_gender'
data["previous_loan_defaults_on_file"] = label1

```

```python
le2 = LabelEncoder()
label2 = le2.fit_transform(data['person_home_ownership'])
data.drop("person_home_ownership", axis=1, inplace=True)

data["person_home_ownership"] = label2
```

```python
le3 = LabelEncoder()
label3 = le3.fit_transform(data['person_education'])
data.drop("person_education", axis=1, inplace=True)

data["person_education"] = label3
```

```python
le4 = LabelEncoder()
label4 = le4.fit_transform(data['loan_intent'])
data.drop("loan_intent", axis=1, inplace=True)

data["loan_intent"] = label4
data
```

```python
x=data[['person_age','person_income','person_emp_exp','loan_amnt','loan_int_rate','loan_percent_income','cb_person_cred_hist_length','credit_score','person_gender','previous_loan_defaults_on_file','person_home_ownership','person_education','loan_intent']]
```

```python
y=data['loan_status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Initialize the Logistic Regression model
model = LogisticRegression()

# Fit the model on the training data
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

print("\nPredictions:")
print(y_pred)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy:", accuracy)
```

```python
print("Confusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
